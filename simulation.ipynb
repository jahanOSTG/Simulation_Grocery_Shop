# Time Between Arrival

```python
data = {
    "Customer no": [1, 2, 3, 4, 5, 6],
    "Random digit": [None, 161 ,658, 359,606,198]
}

cust_df = pd.DataFrame(data)
```

```python
def determine_iat(random_digit, iat_table):
    if pd.isna(random_digit):
        return None

    for _, row in iat_table.iterrows():
        if random_digit <= row["Upper"]:
            return row["IAT"]

    return None


```

```python
cust_df["Time Between Arrival"] = cust_df["Random digit"].apply(lambda x: determine_iat(x, iat_df))
iat_results_df = cust_df[["Customer no","Random digit", "Time Between Arrival"]]
display(iat_results_df)
```

# service between time

```python
data = {
    "Customer no": [1, 2, 3, 4, 5, 6],
    "Random digit": [66,97,89,2,16,70]
}

cust_df = pd.DataFrame(data)
```

```python
def determine_service(random_digit, service_time_table):
    for _, row in service_time_table.iterrows():
        if random_digit <= row["Upper"]:
            return row["Service Time"]
    return None
```

```python
cust_df["Service Between Arrival"] = cust_df["Random digit"].apply(lambda x: determine_service(x, service_df))
service_results_df = cust_df[["Customer no","Random digit", "Service Between Arrival"]]
service_results_df = service_results_df.rename(columns={"Random digit": "Random digit (Service)"})
display(service_results_df)
```

# Simulation Table

```python
simulation_table = pd.merge(
    iat_results_df,
    service_results_df[["Customer no", "Random digit (Service)", "Service Between Arrival"]],
    on="Customer no",
    how="left"
)

print("Merged Simulation Table:")
display(simulation_table)
```

```python
import pandas as pd


cust_df = simulation_table.copy()
cust_df = cust_df.rename(columns={
    "Customer no": "Customer",
    "Time Between Arrival": "IAT",
    "Service Between Arrival": "ST"
})

# The first customer's IAT is typically 0
cust_df.loc[0, "IAT"] = 0

# Convert IAT and ST to integer type if they are float
cust_df["IAT"] = cust_df["IAT"].astype(int)
cust_df["ST"] = cust_df["ST"].astype(int)

cust_df["AT"] = 0
cust_df["TSB"] = 0
cust_df["TSE"] = 0
cust_df["WT"] = 0
cust_df["TTS"] = 0
cust_df["Idle Time"] = 0


for i in range(len(cust_df)):

    if i == 0:
        cust_df.loc[i, "AT"] = 0
    else:
        cust_df.loc[i, "AT"] = cust_df.loc[i-1, "AT"] + cust_df.loc[i, "IAT"]

    if i == 0:
        cust_df.loc[i, "TSB"] = 0
    else:
        cust_df.loc[i, "TSB"] = max(cust_df.loc[i, "AT"], cust_df.loc[i-1, "TSE"])

    cust_df.loc[i, "TSE"] = cust_df.loc[i, "TSB"] + cust_df.loc[i, "ST"]


    cust_df.loc[i, "WT"] = cust_df.loc[i, "TSB"] - cust_df.loc[i, "AT"]


    cust_df.loc[i, "TTS"] = cust_df.loc[i, "WT"] + cust_df.loc[i, "ST"]


    if i == 0:
        cust_df.loc[i, "Idle Time"] = cust_df.loc[i, "TSB"]
    else:
        cust_df.loc[i, "Idle Time"] = max(0, cust_df.loc[i, "TSB"] - cust_df.loc[i-1, "TSE"])


final_df = cust_df[["Customer","IAT","AT","ST","TSB","TSE","WT","TTS","Idle Time"]]
display(final_df)
```
